## Create new dataframe and check State and total no.of cases

```python
cases_df=pd.DataFrame(df['State'].value_counts().reset_index().values,columns=["State","Cases"])
```

```python
cases_df
```

```python
plt.figure(figsize=(16,8))
plt.bar(cases_df['State'],cases_df['Cases'])
plt.title('security breaches in us states between 1 jan 2021 and 8 mar 2023',fontsize=25)
plt.xlabel('States',fontsize=20)
plt.ylabel('no.of breaches',fontsize=20)
plt.show
```

## Which state has maximum cases

```python
cases_df[cases_df['Cases']==cases_df['Cases'].max()]
```

## which state has minimum cases

```python
cases_df[cases_df['Cases']==cases_df['Cases'].min()]
```
